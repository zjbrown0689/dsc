```python
import pandas as pd
import numpy as np
import os
from datetime import datetime, timedelta
```

```python
os.getcwd()
```

```python
users = pd.read_csv("takehome_users.csv", index_col='object_id', encoding='latin-1')
```

```python
users.head()
```

```python
engagement = pd.read_csv('takehome_user_engagement.csv')
```

```python
engagement.head()
```

```python
merged = engagement.join(users, on='user_id', how='inner')
```

```python
merged.head()
```

```python
merged.dtypes
```

```python
merged['time_stamp'] = pd.to_datetime(merged['time_stamp'])
```

```python
merged.dtypes
```

```python
merged['creation_time'] = pd.to_datetime(merged['creation_time'])
```

```python
merged['time_stamp'] = merged['time_stamp'].dt.floor('d').astype(np.int64)
dropped_dups = merged.sort_values(['user_id', 'time_stamp']).drop_duplicates()
```

```python
roll = dropped_dups.groupby('user_id')['time_stamp'].rolling(window=3)
delta = pd.to_timedelta((roll.max()- roll.min())).dt.days
```

```python
delta
```

```python
adopted = delta[delta < 8].index.get_level_values('user_id').tolist()
```

```python
adopted
```

```python
unique_adopted = []
for i in adopted:
    if i not in unique_adopted:
        unique_adopted.append(int(i))
```

```python
unique_adopted
```

```python
merged['user_id'].unique()
```

```python
merged['adopted'] = True
```

```python
for user in merged['user_id'].unique():
    if user in unique_adopted:
        merged.loc[merged['user_id'] == user, 'adopted'] = True
    else:
        merged.loc[merged['user_id'] == user, 'adopted'] = False
```

```python
merged['adopted'].value_counts()
```

```python
dummied = pd.get_dummies(data = merged, columns=['creation_source', 'org_id'], drop_first=True)
```

```python
dummied.columns.to_list()
```

```python
working = dummied.drop(columns=['time_stamp', 'user_id', 'creation_time', 'name', 'email', 'last_session_creation_time', 'invited_by_user_id'])
```

```python
X = working.drop(columns=['adopted'])
y = working[['adopted']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
scaler.transform(X_train)
scaler.transform(X_test)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(strategy='most_frequent')
imp.fit(X_train)
imp.transform(X_train)
imp.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
```

```python
parameters = {'max_depth': [3, 5, 7, 10], 'n_estimators': [50, 100, 200]}

rfc = RandomForestClassifier(random_state=15)
gscv = GridSearchCV(rfc, parameters, scoring = 'f1')
gscv.fit(X_train, y_train)
gscv.best_score_
```

```python
gscv.best_params_
```

```python
rand = RandomForestClassifier(max_depth = 3, n_estimators = 50, random_state = 15)
rand.fit(X_train, y_train)
y_pred = rand.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
imp = pd.DataFrame({'feature':X_train.columns, 'importance':rand.feature_importances_})
```
