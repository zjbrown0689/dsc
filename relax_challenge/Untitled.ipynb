```python
for person in merged['user_id'].unique():
    subset = merged[merged['user_id'] == person]
    for i in range(len(subset)):
        if any(len(subset.between_time(subset.iloc[i].index,\
                                   subset.iloc[i].index + timedelta(days=7), inclusive='both') >= 3)):
            merged['adopted_user'] == True
        else: merged['adopted_user'] == False
```
